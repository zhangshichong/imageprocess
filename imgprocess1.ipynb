图像分为二值图、灰度图、RGB图。
二值图中的像素要么为白色（255），要么为黑色（0）；
灰度图除了白色、黑色还有灰色，把灰度划分为256个等级。
二值图像转为灰度，大于127设置为255，否则为0。
将彩色图像转换为灰度图是图像处理的最基本预处理操作，通常包括下面几种方法：
(1) 浮点算法：Gray=R0.3+G0.59+B0.11
(2) 整数方法：Gray=(R30+G59+B11)/100
(3) 移位方法：Gray=(R28+G151+B77)>>8;
(4) 平均值法：Gray=（R+G+B）/3;(此程序采用算法)
(5) 仅取绿色：Gray=G；
(6) 加权平均值算法：根据光的亮度特性，公式:  R=G=B=R0.299+G*0.587+B0.144
通过上述任一种方法求得Gray后，将原来的RGB(R,G,B)中的R,G,B统一用Gray替换，形成新的颜色RGB(Gray,Gray,Gray)，用它替换原来的RGB(R,G,B)就是灰度图了。改变象素矩阵的RGB值，来达到彩色图转变为灰度图。


```
import cv2 as cv
import matplotlib.pyplot as plt
```

读取图像

```
img = cv.imread('me.jpg', cv.IMREAD_UNCHANGED)
plt.imshow(img)
```

提取某一点的像素

```
test = img[0,0]
print(test)
```

更改像素值


255是白色

```
c = [255, 255, 255]
img[1:100,1:100] = c
plt.imshow(img)
```

读取RGB的每个通道

```
re = img[78, 78]
print(re)
r = img[78, 78, 0]
g = img[78, 78, 1]
b = img[78, 78, 2]
print("r:"+ str(r))
print("g:"+ str(g))
print("b:"+ str(b))
```

使用numpy读取像素

```
img = cv.imread('me.jpg', cv.IMREAD_UNCHANGED)
a = img.item(78, 100, 0)
b = img.item(78, 100, 1)
c = img.item(78, 100, 2)
print(a, b, c)
```

使用numpy修改像素

```
img.itemset((78, 100, 0), 255)
img.itemset((78, 100, 1), 255)
img.itemset((78, 100, 2), 255)
plt.imshow(img)
```
