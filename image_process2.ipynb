```
import cv2 as cv
import matplotlib.pyplot as plt
img = cv.imread('me.jpg')
print(img.shape)#图像形状
print(img.size)#图像像素数量
print(img.dtype)#图像类型
```

ROI
region of interest感兴趣的区域：使用slide进行处理。

通道处理

b [:, :, 0]

g [:, :, 1]

r  [:, :, 2]


```
#使用split函数拆分通道
b, g, r = cv.split(img)
plt.imshow(b)
```

```
#通道合并
new_img = cv.merge([b, g, r])
plt.imshow(new_img)
```
